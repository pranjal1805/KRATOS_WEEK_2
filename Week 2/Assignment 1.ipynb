```python
import numpy as np
import cv2
import matplotlib.pyplot as plt


def main():
    #This captures the image from the live cam
    cap = cv2.VideoCapture(0)
    
    if cap.isOpened():
        ret, frame = cap.read()
        print(ret)
        print(frame)
    else:
        ret = False
    
    
    img1 = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    img2 = cv2.cvtColor(frame, cv2.COLOR_BGR2BGRA)
    img3 = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    img4 = cv2.cvtColor(frame, cv2.COLOR_BGR2HLS)
    img5 = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    img6 = cv2.cvtColor(frame, cv2.COLOR_BGR2XYZ)
    
    #This inverts the image:(looks like the negative)
    img1 = cv2.bitwise_not(img1)
    plt.imshow(img1)
    plt.title('Color Image Inverted')
    plt.xticks([])
    plt.yticks([])
    plt.show
    
    #This converts image from RGB to BGR(When Open CV captures the image it already is in the BGR format)
    img2 = cv2.bitwise_not(img2)
    plt.imshow(img2)
    plt.title('Color Image BGR')
    plt.xticks([])
    plt.yticks([])
    plt.show
    
    #This converts image from BGR to HSV(looks yellowish)
    img3 = cv2.bitwise_not(img3)
    plt.imshow(img3)
    plt.title('Color Image HSV')
    plt.xticks([])
    plt.yticks([])
    plt.show
    
    #This converts image from BGR to HLS(the pic looks kind of pink/purple now!)
    img4 = cv2.bitwise_not(img4)
    plt.imshow(img4)
    plt.title('Color Image HSV')
    plt.xticks([])
    plt.yticks([])
    plt.show
    
    #This converts image from BGR to Gray(the pic looks gray now!)
    img5 = cv2.bitwise_not(img5)
    plt.imshow(img5)
    plt.title('Color Image Gray')
    plt.xticks([])
    plt.yticks([])
    plt.show
    
    #This removed the color from the pic and it looks black and white for some reason
    img6 = cv2.bitwise_not(img6)
    plt.imshow(img6)
    plt.title('Color Image XYZ')
    plt.xticks([])
    plt.yticks([])
    plt.show
    
    cap.release()
    
if __name__ == "__main__":
    main()
```
